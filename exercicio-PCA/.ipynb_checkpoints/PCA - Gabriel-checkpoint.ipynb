```python
# Construindo pipelines com PCA


df = pd.read_csv('/home/gabrielmartins/Aprendizagem-estatistica-classificacao/SAheart.data')

X = df.drop(['chd'], axis=1)
y = df['chd']
le = preprocessing.LabelEncoder()
X['famhist'] = le.fit_transform(X['famhist'])


pca = PCA(n_components = 3)
pca.fit(X)
X_pca = pca.transform(X)

X_train_pca, X_test_pca, y_train_pca, y_test_pca = train_test_split(X_pca,y, test_size=0.3 , random_state = 101)

pipe_lr = Pipeline( [ ('scl', StandardScaler()), ('clf', LogisticRegression()) ] )
pipe_knn = Pipeline([('scl', StandardScaler()), ('clf', KNeighborsClassifier())])
pipe_dt = Pipeline([('scl', StandardScaler()), ('clf', DecisionTreeClassifier())])
pip_rand = Pipeline([('scl', StandardScaler()), ('clf', RandomForestClassifier())])

# Lista de pipelines a serem executados
pipelines = [pipe_lr, pipe_knn, pipe_dt, pip_rand]

# Dicionário para facilitar identificacao
pipe_dict = {0: 'Logistic Regression', 1: 'KNN', 2: 'Decision Tree', 3: 'Random Forest'}

# aplicando fit
# Generaliza a execucao do fit de cada ultima funcao do pipe
for pipe in pipelines:
    pipe.fit(X_train_pca, y_train_pca)

# Compara acurácia
for idx, val in enumerate(pipelines):
    print('%s pipeline test accuracy with PCA: %.3f' % (pipe_dict[idx], val.score(X_test_pca, y_test_pca)))

# para cada modelo treinado obtem val score
best_acc = 0.0
best_clf = 0
best_pipe = ''
for idx, val in enumerate(pipelines):
    # Descobre o melhor val.score e armazen em best_clf
    if val.score(X_test_pca, y_test_pca) > best_acc:
        best_acc = val.score(X_test_pca, y_test_pca)
        best_pipe = val
        best_clf = idx
print('Classifier with best accuracy: %s' % pipe_dict[best_clf])
```
